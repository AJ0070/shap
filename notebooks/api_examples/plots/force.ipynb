# `force` plot

This notebook demonstrates how to use `shap.plots.force` with the modern `Explanation` interface and the matplotlib `ax` API.

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

import shap

# Train a small model
X, y = shap.datasets.california(n_points=300)
model = RandomForestRegressor(n_estimators=30, random_state=0)
model.fit(X, y)

# Compute SHAP values
explainer = shap.TreeExplainer(model)
explanation = explainer(X)
```

```python
# Initialize SHAP JavaScript for interactive force plots in notebooks.
shap.initjs()
```

## Basic force plot

Pass a single row of an `Explanation` to visualize a local explanation.

```python
shap.plots.force(explanation[0], matplotlib=True)
```

## Customize with explicit axes

The force plot supports matplotlib's explicit object-oriented API by accepting an `ax` argument and returning that axis when `show=False`.

```python
fig, ax = plt.subplots(figsize=(10, 2.5))
returned_ax = shap.plots.force(explanation[1], matplotlib=True, show=False, ax=ax)
returned_ax.set_title("Force Plot on User-Provided Axes")
plt.show()
```

## JavaScript force plot

For notebooks that support JavaScript outputs, `matplotlib=False` returns the interactive force visualizer.

```python
shap.plots.force(explanation[2])
```

# `force` plot

This notebook documents the `shap.plots.force` API, including the explicit matplotlib Axes interface.

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestRegressor

import shap

# Build a small tabular regression setup with deterministic behavior.
diabetes = load_diabetes()
X = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y = diabetes.target

model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X, y)

explainer = shap.Explainer(model, X)
shap_values = explainer(X.iloc[:30], check_additivity=False)
```

## Standard interactive force plot

`shap.plots.force` can be called directly with an `Explanation` object.

```python
shap.plots.force(shap_values[0])
```

## Matplotlib mode with explicit Axes

Pass `ax` and set `show=False` to draw onto an existing Axes and get it back for further customization.

```python
fig, ax = plt.subplots(figsize=(12, 3))
returned_ax = shap.plots.force(shap_values[0], matplotlib=True, ax=ax, show=False)
assert returned_ax is ax
ax.set_title("Force plot on a provided Axes")
plt.tight_layout()
plt.show()
```

## Uses current figure when `ax=None`

When no Axes is supplied, the plot uses the current figure/axes from matplotlib state.

```python
fig = plt.figure(figsize=(12, 3))
returned_ax = shap.plots.force(shap_values[1], matplotlib=True, show=False)
assert returned_ax.figure is fig
returned_ax.set_title("Force plot using current figure")
plt.tight_layout()
plt.show()
```
